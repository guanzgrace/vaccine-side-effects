```python
with open('df_unified_20211009.pkl', 'rb') as f:
    df = pkl.load(f)
```

```python
def age1855(x):
    if x <= 55:
        return "yes"
    return "no"

def age55up(x):
    if x>55:
        return "yes"
    return "no"
```

```python
df['age1855'] = df["age"].apply(age1855)
df['age55up'] = df["age"].apply(age55up)
```

```python
df['healthy'].describe()
```

```python
def descriptive_statistics(participants):
    this_df = df.where(df["participant_num"].isin(participants)).dropna(how='all')
    n = len(list(set(this_df["participant_num"].values)))
    
    df_no_dupes = this_df.drop_duplicates(subset=["participant_num"])
    gender_describe = df_no_dupes["gender"].describe()
    #print(gender_describe)
    assert gender_describe["count"] == n
    
    age_describe = df_no_dupes["age"].describe()
    age_describe_1855 = df_no_dupes["age1855"].describe()
    age_describe_55up = df_no_dupes["age55up"].describe()
    #print(age_describe)
    #print(age_describe_1)
    
    umc_describe = df_no_dupes["healthy"].describe()
    #print(umc_describe)
    
    print("n = %d" % (n))
    
    #Gender
    g_count = gender_describe["count"]
    g_freq = gender_describe["freq"]
    if gender_describe["top"] == "male":
        print("Female (pct) %d (%.1f%%)" % (g_count - g_freq, 100 * (g_count - g_freq) / g_count))
        print("Male (pct) %d (%.1f%%)" % (g_freq, 100 * g_freq / g_count))
    else:
        print("Female (pct) %d (%.1f%%)" % (g_freq, 100 * g_freq / g_count))
        print("Male (pct) %d (%.1f%%)" % (g_count - g_freq, 100 * (g_count - g_freq) / g_count))
        
    # Age
    print("Age (std) (range) %.2f (%.2f) [%d-%d] median %.2f" % (age_describe["mean"], age_describe["std"],
                                                                 age_describe["min"], age_describe["max"],
                                                                 age_describe["50%"]))
    
    # Age 18-55 
    g_count = age_describe_1855["count"]
    g_freq = age_describe_1855["freq"]
    if age_describe_1855["top"] == "no":
        print("18-55 (pct) %d (%.1f%%)" % (g_count - g_freq, 100 * (g_count - g_freq) / g_count))
    else:
        print("18-55 (pct) %d (%.1f%%)" % (g_freq, 100 * g_freq / g_count))
        
    # Age >55
    g_count = age_describe_55up["count"]
    g_freq = age_describe_55up["freq"]
    if age_describe_55up["top"] == "no":
        print(">55 (pct) %d (%.1f%%)" % (g_count - g_freq, 100 * (g_count - g_freq) / g_count))
    else:
        print(">55 (pct) %d (%.1f%%)" % (g_freq, 100 * g_freq / g_count))
        
    # Underlying medical condition
    g_count = umc_describe["count"]
    g_freq = umc_describe["freq"]
    if umc_describe["top"] == True:
        print("Underlying Medical Condition (pct) %d (%.1f%%)" % (g_count - g_freq, 100 * (g_count - g_freq) / g_count))
        print("No Underlying Medical Condition (pct) %d (%.1f%%)" % (g_freq, 100 * g_freq / g_count))
    else:
        print("Underlying Medical Condition (pct) %d (%.1f%%)" % (g_freq, 100 * g_freq / g_count))
        print("No Underlying Medical Condition (pct) %d (%.1f%%)" % (g_count - g_freq, 100 * (g_count - g_freq) / g_count))
      
```

```python
def descriptive_statistics_copy_text(participants):
    this_df = df.where(df["participant_num"].isin(participants)).dropna(how='all')
    n = len(list(set(this_df["participant_num"].values)))
    
    df_no_dupes = this_df.drop_duplicates(subset=["participant_num"])
    gender_describe = df_no_dupes["gender"].describe()
    #print(gender_describe)
    assert gender_describe["count"] == n
    
    age_describe = df_no_dupes["age"].describe()
    age_describe_1855 = df_no_dupes["age1855"].describe()
    age_describe_55up = df_no_dupes["age55up"].describe()
    #print(age_describe)
    #print(age_describe_1)
    
    umc_describe = df_no_dupes["healthy"].describe()
    #print(umc_describe)
    
    print("n = %d" % (n))
    
    #Gender
    g_count = gender_describe["count"]
    g_freq = gender_describe["freq"]
    if gender_describe["top"] == "male":
        print("%d (%.1f%%)" % (g_count - g_freq, 100 * (g_count - g_freq) / g_count))
        print("%d (%.1f%%)" % (g_freq, 100 * g_freq / g_count))
    else:
        print("%d (%.1f%%)" % (g_freq, 100 * g_freq / g_count))
        print("%d (%.1f%%)" % (g_count - g_freq, 100 * (g_count - g_freq) / g_count))
        
    # Age
    print("%.2f (%.2f) [%d-%d] %.2f" % (age_describe["mean"], age_describe["std"],
                                                                 age_describe["min"], age_describe["max"],
                                                                 age_describe["50%"]))
    
    
    print("%d" % (n))
    
    # Age 18-55 
    g_count = age_describe_1855["count"]
    g_freq = age_describe_1855["freq"]
    if age_describe_1855["top"] == "no":
        print("%d (%.1f%%)" % (g_count - g_freq, 100 * (g_count - g_freq) / g_count))
    else:
        print("%d (%.1f%%)" % (g_freq, 100 * g_freq / g_count))
        
    # Age >55
    g_count = age_describe_55up["count"]
    g_freq = age_describe_55up["freq"]
    if age_describe_55up["top"] == "no":
        print("%d (%.1f%%)" % (g_count - g_freq, 100 * (g_count - g_freq) / g_count))
    else:
        print("%d (%.1f%%)" % (g_freq, 100 * g_freq / g_count))
        
        
    # Underlying medical condition
    g_count = umc_describe["count"]
    g_freq = umc_describe["freq"]
    if umc_describe["top"] == True:
        print("%d (%.1f%%)" % (g_count - g_freq, 100 * (g_count - g_freq) / g_count))
        print("%d (%.1f%%)" % (g_freq, 100 * g_freq / g_count))
    else:
        print("%d (%.1f%%)" % (g_freq, 100 * g_freq / g_count))
        print("%d (%.1f%%)" % (g_count - g_freq, 100 * (g_count - g_freq) / g_count))
      
```
